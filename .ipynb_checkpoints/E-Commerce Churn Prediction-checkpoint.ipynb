```python
import pandas as pd

ecommerce_df = pd.read_excel("ecommerce_dataset.xlsx")
```

```python
ecommerce_df = ecommerce_df.drop(columns={
    'Unnamed: 0', 'Unnamed: 1'
})
```

```python
ecommerce_df = ecommerce_df.drop(ecommerce_df.index[0])
```
