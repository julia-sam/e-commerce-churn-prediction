```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in df1.columns:
    counts = df1[col].value_counts()
```

```python
plt.figure(figsize=(15,40))
plot_num = 1

for col in df1.columns:
    plt.subplot(10,2,plot_num)
    sns.countplot(data=df1, x=col)
    plot_num+=1
    plt.tight_layout()
```

```python
grouped_df_ordercat = df1.groupby(['PreferedOrderCat', 'Gender']).size().reset_index(name='Count')
grouped_df_ordercat
```

```python
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(data=grouped_df_ordercat, x='PreferedOrderCat', y='Count', hue='Gender')

plt.title('Comparison of Order Categories by Gender')
plt.xlabel('Order Category')
plt.ylabel('Count of Orders')

plt.legend(title='Gender')
plt.show()
```

```python
grouped_df_churn_gender = df1.groupby(['Churn', 'Gender']).size().reset_index(name='Count')
grouped_df_churn_gender
```

```python
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(data=grouped_df_churn_gender, x='Churn', y='Count', hue='Gender')

plt.title('Comparison of Churn by Gender')
plt.xlabel('Churn')
plt.ylabel('Counts')

plt.legend(title='Gender')
plt.show()
```

```python
grouped_df_churn_complaints = df1.groupby(['Complain', 'Churn']).size().reset_index(name='Count')
grouped_df_churn_complaints
```

```python
df1['Complain'] = df1['Complain'].astype(str)
df1['Churn'] = df1['Churn'].astype(str)

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(data=grouped_df_churn_complaints, x='Complain', y='Count', hue='Churn')

plt.title('Effects of Complaints on Customer Churn')
plt.xlabel('Complained')
plt.ylabel('Count of Customers')

plt.legend(title='Churn')
plt.show()
```

```python
grouped_df_churn_score = df1.groupby(['Churn', 'SatisfactionScore']).size().reset_index(name='Count')
grouped_df_churn_score
```

```python
df1['Complain'] = df1['Complain'].astype(str)
df1['Churn'] = df1['Churn'].astype(str)

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(data=grouped_df_churn_score, x='SatisfactionScore', y='Count', hue='Churn')

plt.title('Effects of Satisfaction Score on Customer Churn')
plt.xlabel('Satisfaction Score')
plt.ylabel('Count of Customers')

plt.legend(title='Churn')
plt.show()
```

# Product Popularity

```python
popular_products = pd.DataFrame(df1.groupby('PreferedOrderCat')['SatisfactionScore'].value_counts().reset_index(name='count'))
most_popular = popular_products.sort_values('SatisfactionScore', ascending=False)
most_popular.head(10)
```

```python
popular_products['SatisfactionScore'] = popular_products['SatisfactionScore'].astype(str)
popular_products = popular_products.sort_values(by='SatisfactionScore')

plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(data=popular_products, x='PreferedOrderCat', y='count', hue='SatisfactionScore')

plt.title('Product Popularity by Satisfaction Score')
plt.xlabel('Preferred Order Category')
plt.ylabel('Count')
plt.legend(title='Satisfaction Score')

plt.xticks(rotation=45)
plt.show()
```

```python
marital_status_product_satisfaction = df1.groupby('PreferedOrderCat')['MaritalStatus'].value_counts()
marital_status_product_satisfaction
```

```python
marital_status_product_satisfaction = df1.groupby('PreferedOrderCat')['MaritalStatus'].value_counts().reset_index(name='Count')
total_counts = marital_status_product_satisfaction.groupby('PreferedOrderCat')['Count'].transform('sum')

marital_status_product_satisfaction['Proportion'] = marital_status_product_satisfaction['Count'] / total_counts

plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(data=marital_status_product_satisfaction, x='PreferedOrderCat', y='Proportion', hue='MaritalStatus')

plt.title('Proportion of Each Marital Status in Each Preferred Order Category')
plt.xlabel('Preferred Order Category')
plt.ylabel('Proportion')
plt.legend(title='Marital Status')

plt.xticks(rotation=45)
plt.show()
```
